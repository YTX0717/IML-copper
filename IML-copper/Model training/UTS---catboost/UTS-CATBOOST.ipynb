```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from catboost import CatBoostRegressor, Pool
import optuna
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.model_selection import cross_val_score

# Data loading and feature engineering
df4 = pd.read_excel('calculated_properties_EC.xlsx')
df4['TT'] = df4['T1'] * df4['T2']
df4['RR'] = df4['R2'] + df4['R1']
X = df4[['B3', 'E6', 'S1_1','TT','RR']]
y = df4['UTS']

# Split dataset
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=24)

# Define CatBoost optimization objective function
def objective(trial):
    params = {
        'iterations': trial.suggest_int('iterations', 500, 1500),  # Number of iterations
        'learning_rate': trial.suggest_float('learning_rate', 0.005, 0.2, log=True),  # Learning rate
        'depth': trial.suggest_int('depth', 5, 8),  # Tree depth
        'l2_leaf_reg': trial.suggest_float('l2_leaf_reg', 1, 5),  # L2 regularization
        'random_strength': trial.suggest_float('random_strength', 1e-9, 10, log=True),  # Random strength
        'bagging_temperature': trial.suggest_float('bagging_temperature', 0.0, 1.0),  # Bagging temperature
        'border_count': trial.suggest_int('border_count', 32, 255),  # Number of bins
        'verbose': False,
        # Additional optimization parameters:
        'grow_policy': trial.suggest_categorical('grow_policy', ['SymmetricTree', 'Depthwise']),  # Symmetric tree and depth-first
        'min_data_in_leaf': trial.suggest_int('min_data_in_leaf', 1, 50),  # Minimum data in leaf node
        # 'one_hot_max_size': trial.suggest_int('one_hot_max_size', 2, 10)  # Maximum features for one-hot encoding
    }
    
    model = CatBoostRegressor(**params)
    
    try:
        score = -np.mean(cross_val_score(
            model,
            X_train,
            y_train,
            cv=5,
            scoring='neg_mean_squared_error',
            n_jobs=-1,
            error_score='raise'
        ))
        return score
    except Exception as e:
        print(f"Parameter combination failed: {str(e)}")
        return np.inf

# Execute optimization
study = optuna.create_study(direction='minimize')
study.optimize(objective, n_trials=30, show_progress_bar=True)

# Train final model
best_params = study.best_params
best_model = CatBoostRegressor(**best_params)
best_model.fit(X_train, y_train, verbose=100)  # Show training progress

# Prediction and evaluation
def evaluate(model, X, y, set_name):
    y_pred = model.predict(X)
    print(f"\n{set_name} evaluation metrics:")
    print(f"MSE: {mean_squared_error(y, y_pred):.4f}")
    print(f"RMSE: {np.sqrt(mean_squared_error(y, y_pred)):.4f}")
    print(f"MAE: {mean_absolute_error(y, y_pred):.4f}")
    print(f"R²: {r2_score(y, y_pred):.4f}")

evaluate(best_model, X_train, y_train, "Training set")
evaluate(best_model, X_test, y_test, "Test set")

# Feature importance analysis
print("\nFeature importance values:")
feature_importance = best_model.get_feature_importance()
for name, score in zip(X.columns, feature_importance):
    print(f"{name}: {score:.4f}")

# Feature importance visualization
plt.figure(figsize=(10, 6))
sorted_idx = np.argsort(feature_importance)
plt.barh(range(len(sorted_idx)), feature_importance[sorted_idx], align='center')
plt.yticks(range(len(sorted_idx)), X.columns[sorted_idx])
plt.title("CatBoost Feature Importance")
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import optuna
from sklearn.base import BaseEstimator, RegressorMixin
from sklearn.model_selection import cross_val_score, StratifiedKFold
from xgboost import plot_importance
import xgboost as xgb
import numpy as np
from xgb_wrapper import XGBWrapper
from sklearn.metrics import make_scorer, mean_squared_error, mean_absolute_error
import pandas as pd

# Data loading and feature engineering
df4 = pd.read_excel('calculated_properties_EC.xlsx')
df4['TT'] = df4['T1'] * df4['T2']
df4['RR'] = df4['R2'] + df4['R1']
X = df4[['B3', 'B5', 'E6', 'S1_1','TT','RR']]
y = df4['UTS']

# Split dataset
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=24)

# Define evaluation metrics
scoring = {
    'r2': 'r2',  # R²
    'mse': make_scorer(mean_squared_error),  # Mean Squared Error
    'rmse': make_scorer(lambda y_true, y_pred: np.sqrt(mean_squared_error(y_true, y_pred))),  # Root Mean Squared Error
    'mae': make_scorer(mean_absolute_error)  # Mean Absolute Error
}

# Use Stratified K-Fold cross-validation
# Bin the target variable to enable stratification
y_binned = pd.cut(y, bins=10, labels=False)  # Divide target variable into 10 intervals
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=24)

# Perform stratified five-fold cross-validation
cv_results = {}
for metric_name, metric_scorer in scoring.items():
    scores = cross_val_score(best_model, X, y, cv=skf.split(X, y_binned), scoring=metric_scorer)
    cv_results[metric_name] = scores

# Output cross-validation results
for metric_name, scores in cv_results.items():
    print(f"Five-fold cross-validation {metric_name.upper()} scores:")
    print(scores)
    print(f"Average {metric_name.upper()}: {scores.mean():.4f} (±{scores.std():.4f})\n")

# Split training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=24)

# Prediction and evaluation
y_pred_train = best_model.predict(X_train)
y_pred_test = best_model.predict(X_test)

# Convert to lists
y_pred_train_list = y_pred_train.tolist()
y_pred_test_list = y_pred_test.tolist()

from sklearn import metrics

# Calculate metrics
print("Training set evaluation metrics:")
print(f"MSE: {metrics.mean_squared_error(y_train, y_pred_train_list):.4f}")
print(f"RMSE: {np.sqrt(metrics.mean_squared_error(y_train, y_pred_train_list)):.4f}")
print(f"MAE: {metrics.mean_absolute_error(y_train, y_pred_train_list):.4f}")
print(f"R²: {metrics.r2_score(y_train, y_pred_train_list):.4f}")

print("\nTest set evaluation metrics:")
print(f"MSE: {metrics.mean_squared_error(y_test, y_pred_test_list):.4f}")
print(f"RMSE: {np.sqrt(metrics.mean_squared_error(y_test, y_pred_test_list)):.4f}")
print(f"MAE: {metrics.mean_absolute_error(y_test, y_pred_test_list):.4f}")
print(f"R²: {metrics.r2_score(y_test, y_pred_test_list):.4f}")
```
